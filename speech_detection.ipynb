```
from deepspeech import Model
import numpy as np
import os
import wave

from IPython.display import Audio
```

```
acoustic_model_file_path = '/content/deepspeech-0.9.3-models.pbmm'
language_model_file_path = '/content/deepspeech-0.9.3-models.scorer'
beam_width = 500
language_model_alpha = 0.93
language_model_beta = 1.18

model = Model(acoustic_model_file_path)
model.enableExternalScorer(language_model_file_path)
```

```
model.setScorerAlphaBeta(language_model_alpha, language_model_beta)
model.beamWidth(beam_width)
```

```

def read_wav_file(filename):
    with wave.open(filename, 'rb') as w:
        rate = w.getframerate()
        frames = w.getnframes()
        buffer = w.readframes(frames)
        print("Rate:", rate)
        print("Frames:", frames)
        print("Buffer Len:", len(buffer))

    return buffer, rate
```

```
def transcribe_batch(audio_file):
    buffer, rate = read_wav_file(audio_file)
    data = np.frombuffer(buffer, dtype=np.int16)
    return model.stt(data)
```
